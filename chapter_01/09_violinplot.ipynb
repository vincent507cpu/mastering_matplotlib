前面介绍过了箱型图，它可以反映原始数据大概的分布，但无法展示数据的概率密度。于是小提琴图应运而生。

小提琴图有两个函数，`violin` 接收一个包含统计值的字典；`violinplot` 接收原始数据，自行计算统计值。本文介绍 `violinplot` 。

```python
violin(vpstats, positions=None, vert=True, widths=0.5, showmeans=False, showextrema=True, showmedians=False)

violinplot(dataset, positions=None, vert=True, widths=0.5, showmeans=False, showextrema=True, showmedians=False, quantiles=None, points=100, bw_method=None, *, data=None)
```

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Heiti TC']

np.random.seed(1)
data = [np.random.randint(0, 100) for _ in range(100)]
````

````python
fig, ax = plt.subplots(figsize = (2, 4))
ax.violinplot(data)
ax.set_title('原始图像')
plt.show()
````

# `dataset`，`data`
数据源，不再赘述。

# `positions`
与数据长度相同的数组，默认值为 [1, 2, ..., n]。图像的位置。

````python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (4, 4))

ax1.violinplot(data)
ax1.set_title('positions = [1]')

ax2.violinplot(data, positions = [2])
ax2.set_title('positions = [2]')

fig.tight_layout()
plt.show()
````

# `vert`
布尔值，默认值为 True。为 True 则竖直显示，否则为水平显示。

````python
fig, ax = plt.subplots(figsize = (2, 4))
ax.violinplot(data)
ax.set_title('vert = True')
fig.savefig('pics/1-9-11.png')
plt.show()

fig, ax = plt.subplots(figsize = (4, 2))
ax.violinplot(data, vert = False)
ax.set_title('vert = False')
plt.show()
````

# `widths`
标量或与数据等长的数组，默认值为 0.5。

````python
fig, ax = plt.subplots(figsize = (4, 4))

ax.violinplot([data] * 2, widths = [0.5, 1])
ax.set_title('widths = [0.5, 1]')

plt.show()
````

# `showmeans`
布尔值，默认为 False。为 True 则显示中线。

````python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (4, 4))

ax1.violinplot(data)
ax1.set_title('showmeans = False')

ax2.violinplot(data, showmeans = True)
ax2.set_title('showmeans = True')

fig.tight_layout()
plt.show()
````

# `showextrema`
布尔值，默认为 True。为 True 则显示两根极值线。

````python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (4, 4))

ax1.violinplot(data)
ax1.set_title('showextrema = True')

ax2.violinplot(data, showextrema = False)
ax2.set_title('showextrema = False')

fig.tight_layout()
plt.show()
````

# `showmedians`
布尔值，默认为 False。为 True 则显示中位线。

````python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (4, 4))

ax1.violinplot(data)
ax1.set_title('showmedians = False')

ax2.violinplot(data, showmedians = True)
ax2.set_title('showmedians = True')

fig.tight_layout()
plt.show()
````

# `quantiles`
二元浮点数数组，默认为 [0, 1]。显示上下两个分位数的位置。

````python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (4, 4))

ax1.violinplot(data)
ax1.set_title('quantiles = None')

ax2.violinplot(data, quantiles = [0.4, 0.6])
ax2.set_title('quantiles = [0.4, 0.6]')

fig.tight_layout()
plt.show()
````

# `points`
正整数，默认为 100。定义高斯核函数密度的点数量。

````python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (4, 4))

ax1.violinplot(data)
ax1.set_title('points = 100')

ax2.violinplot(data, points = 10)
ax2.set_title('points = 10')

fig.tight_layout()
plt.show()
````

# `bw_method`
- None：传入 'scott'；
- 字符串：'scott', 'silverman' 中的一个；
- 标量：传入 kde.factor；
- 可调用函数：matplotlib.mlab.GaussianKDE 实体。

计算图形的带宽。

````python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (4, 4))

ax1.violinplot(data)
ax1.set_title('bw_method = None')

ax2.violinplot(data, bw_method = 0.6)
ax2.set_title('bw_method = 0.6')

fig.tight_layout()
plt.show()
````
