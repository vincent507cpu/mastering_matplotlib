前几篇文章介绍了常用的图形，本篇开始介绍一些辅助的图形。本篇从各种直线开始，因为是都是直线，直线的参数在这些线上基本适用，故不再赘述。

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Heiti TC']
````

# `hlines`
画一条或多条与 x 轴平行的水平线段。
```python
hlines(y, xmin, xmax, colors=None, linestyles='solid', label='', *, data=None, **kwargs)
```

````python
y = 1
xmin, xmax = 3, 5

fig, ax = plt.subplots(figsize = (4, 4))

ax.hlines(y, xmin, xmax)
ax.set_title('hlines')

plt.show()
````

## `y`
浮点数或浮点数数组，水平线段的高。

````python
xmin, xmax = 3, 5

fig, (ax1, ax2) = plt.subplots(2, figsize = (4, 8))

y1 = 1
ax1.hlines(y1, xmin, xmax)
ax1.set_title('一条水平线段')

y2 = np.linspace(1, 2, 4)
ax2.hlines(y2, xmin, xmax)
ax2.set_title('4 条水平线段')

fig.tight_layout()
plt.show()
````

## `xmin` & `xmax`
浮点数或浮点数数组。每条线段的起始、终止坐标。若为数组，`y` 也必须是数组且长度相等。

````python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize = (4, 12))

y1, xmin, xmax = 1, 3, 5
ax1.hlines(y1, xmin, xmax)
ax1.set_title('一条水平线段')

y2 = np.linspace(1, 2, 4)
ax2.hlines(y2, xmin, xmax)
ax2.set_title('4 条长度相同的水平线段')

xmin2 = [0, 0.5, 1, 1.5]
xmax2 = [1, 2, 3, 4]
ax3.hlines(y2, xmin2, xmax2)
ax3.set_title('4 条长度不同的水平线段')

fig.tight_layout()
plt.show()
````

## `**kwargs`
接收 `matplotlib.collections.LineCollection` 的参数，不再赘述。

# `vlines` 
画一条或多条与 x 轴平行的垂直线段。
```python
vlines(x, ymin, ymax, colors=None, linestyles='solid', label='', *, data=None, **kwargs)
```
## `x`，`ymin`，`ymax` 
与 `hlines` 里的 `y`，`xmin`， `xmax` 对应，其它参数完全相同，不再赘述。
## `**kwargs`
接受 `matplotlib.lines.Line2D` 的参数，不再赘述。

# `axhline`
画一条以 y 轴对称、一侧长度为 0 ~ 1 的水平线段。
```python
axhline(y=0, xmin=0, xmax=1, **kwargs)
```

## `y`
浮点数，默认为 0。
## `xmin` & `xmax`
浮点数，均需在 0 ~ 1 之间。`xmin` 的默认值为 0，`xmax` 的默认值为 1。线段首尾坐标为 ((xmin - xmax), (xmax - xmin))。

````python
y = 0.5
xmin, xmax = 0.25, 0.75

fig, ax = plt.subplots(figsize = (4, 4))

ax.axhline(y, xmin, xmax)
ax.set_xlim((-1, 1))
ax.set_title('axhline')

plt.show()
````

## `**kwargs`
接受 `matplotlib.lines.Line2D` 参数，不再赘述。

# `axvline`
画一条以 x 轴对称、一侧长度为 0 ~ 1 的垂直线段。
```python
axvline(x=0, ymin=0, ymax=1, **kwargs)
```
## `x` & `ymin` & `ymax` `**kwargs`
与 `axhline` 里的参数对应或相同，不再赘述。

# `axline`
绘制一条直线。提供两点或一个点和斜率确定一条直线。
```python
axline(xy1, xy2=None, *, slope=None, **kwargs)
```
## `xy1` & `xy2`
两个浮点数组成的数组，`xy2` 可以省略。

````python
fig, ax = plt.subplots(figsize = (4, 4))

ax.axline((1, 1), (2, 3))
ax.set_title('经过点 (1, 1) 和 (2, 3) 的直线')
ax.set_xlim((0, 4))
ax.set_ylim((0, 4))
fig.savefig('pics/1-11-5.png')
plt.show()
````

## `slope`
浮点数。直线的斜率，与 `xy2` 有且只能有一个。

````python
fig, ax = plt.subplots(figsize = (4, 4))

ax.axline((1, 1), slope = 2)
ax.set_title('经过点 (1, 1) 斜率为 2 的直线')
ax.set_xlim((0, 4))
ax.set_ylim((0, 4))

plt.show()
````

## `**kwargs`
接受 `matplotlib.lines.Line2D` 的参数，不再赘述。
