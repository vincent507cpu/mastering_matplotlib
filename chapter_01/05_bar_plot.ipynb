很多人混淆了直方图和柱状图，它们的区别是：
- 柱状图比较单一数据之间的大小；
- 直方图统计所有数据的分布。

本文先来讨论柱状图。

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Heiti TC']

df = pd.DataFrame({'var1':[1, 2, 3, 4, 5], 'var2':[10, 20, 30, 40, 50]}, index=['a', 'b', 'c', 'd', 'e'])
df
````

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像')

ax2.set_title('对照')
fig.tight_layout()
plt.show()
````

# 函数格式
```python
bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)
```mm

## `x`, `height`, `data`
`x` 是每根柱的 x 轴刻度，`height` 是对应柱子的高度，`data` 与 `plot` 中的同名参数相似。

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像')

ax2.bar(x = 'var1', height = 'var2', data = df)
ax2.set_title('对照')
fig.tight_layout()
plt.show()
````

## `width`
柱子相对刻度的宽度，默认值是 0.8。

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像，宽度 0.8')

ax2.bar(x = df.index, height = df['var1'], width = 0.5)
ax2.set_title('宽度 0.5')
fig.tight_layout()
plt.show()
````

## `bottom`
柱子底部距离 x 轴的相对高度，默认值是 0。

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像，bottom = 0')

ax2.bar(x = df.index, height = df['var1'], bottom = 1)
ax2.set_title('bottom = 1')
ax2.set_ylim(0, 6)
fig.tight_layout()
plt.show()
````

## `align`
柱子相对于 x 轴刻度的位置，可选择的值是 `center` 和 `edge`（柱子左侧与x 轴刻度对齐m）。

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像，柱子居中')

ax2.bar(x = df.index, height = df['var1'], align = 'edge')
ax2.set_title('柱子左侧与 x 轴刻度对齐')
ax2.set_xlim(-1, 5)
fig.tight_layout()
plt.show()
````

若想让柱子右侧与 x 轴刻度对齐，需设置 `align = 'edge'` 与 `width` 为负。

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像，柱子居中')

ax2.bar(x = df.index, height = df['var1'], align = 'edge', width=-0.8)
ax2.set_title('柱子右侧与 x 轴刻度对齐')
ax2.set_xlim(-1, 5)
fig.tight_layout()
plt.show()
````

# 额外的参数有：
## `color`、`edgecolor`、`linewidth`
这三个参数与之前 `plot` 里同名参数作用相同，不再赘述。
## `tick_label`
每个柱子在坐标轴上显示的名称，默认是 `None`，此时使用 `x` 值。a

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像，柱子居中')

ax2.bar(x = df.index, height = df['var1'], tick_label = (1, 2, 3, 4, 5))
ax2.set_title('tick_label')
ax2.set_xlim(-1, 5)
fig.tight_layout()
plt.show()
````

## `label`
字符串或列表。如果是单一字符串，全部柱子都被这个字符串标记；如果是列表，长度必须与数据长度相等，每个元素对应一个数据a。

````python
fig, (ax1, ax2, ax3) = plt.subplots(3)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像，柱子居中')

ax2.bar(x = df.index, height = df['var1'], label='单一字符串标签')
ax2.set_title('单一字符串标签')
ax2.legend()

ax3.bar(x = df.index, height = df['var1'], label=['a', 'b', 'c', 'd', 'e'])
ax3.set_title('标签列表')
ax3.legend()

fig.tight_layout()
plt.show()
````

## `xerr`, `yerr`
绘制 x 轴、y 轴方向的误差棒。a

````python
fig, (ax1, ax2, ax3) = plt.subplots(3)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像，柱子居中')

ax2.bar(x = df.index, height = df['var1'], xerr=[0.5, 0.7, 0.9, 1.1, 1.3])
ax2.set_title('x 轴误差棒')
ax2.legend()

ax3.bar(x = df.index, height = df['var1'], yerr=[0.5, 0.7, 0.9, 1.1, 1.3])
ax3.set_title('y 轴误差棒')
ax3.legend()

fig.tight_layout()
plt.show()
````

## `ecolor`
误差棒颜色，可以是字符串或列表，默认值为 `black`。

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像，柱子居中')

ax2.bar(x = df.index, height = df['var1'], 
        yerr=[0.5, 0.7, 0.9, 1.1, 1.3], ecolor = ['red', 'green', 'blue', 'yellow', 'black'])
ax2.set_title('y 轴误差棒不同颜色')
ax2.legend()

fig.tight_layout()
plt.show()
````

## `capsize`
误差棒两端冠的尺寸，接受一个浮点数，默认值为 `rcParams["errorbar.capsize"]` 值。

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var1'])
ax1.set_title('原始图像，柱子居中')

ax2.bar(x = df.index, height = df['var1'], 
        yerr=[0.5, 0.7, 0.9, 1.1, 1.3], capsize = 10.0)
ax2.set_title('误差棒冠的长度：2.0')
ax2.legend()

fig.tight_layout()
plt.show()
````

## `error_kw`
误差棒的参数，接受 `ax.errorbar` 参数，请查阅官方文档。在此定义的参数将覆盖 `bar` 里的参数。

## `log`
布尔值，若为 `True`，则将 y 轴进行对数转换。

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(x = df.index, height = df['var2'])
ax1.set_title('原始图像，柱子居中')

ax2.bar(x = df.index, height = df['var2'], log = True)
ax2.set_title('y 轴误差棒不同颜色')
ax2.legend()

fig.tight_layout()
plt.show()
````

此外也可以传入 `matplotlib.patches.Rectangle` 的参数，有需要会提到。
