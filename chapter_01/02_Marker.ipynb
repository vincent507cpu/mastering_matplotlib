标记（marker）用来在图上表示数据点的位置、类型，用于折线图和散点图中标记数据点。

在 `ax.plot` 和 `ax.scatter` 里与标记有关的参数有：
- `fmt`：为数据点添加标记，控制线的颜色、款式
- `marker`：数据点形状
- `markersize`：数据点大小
- `markeredgewidth` 或 `mec`：数据点边宽
- `markeredgecolor` 或 `mew`：数据点边的颜色
- `markerfacecolor` 或 `mfc`：数据点主要颜色
- `markerfacecoloralt` 或 `mfcalt`：在有渐变色的标记中作为次要颜色

```python
markers = ['.', ',', 'o', 'v', '^', 
           '<', '>', '1', '2', '3', 
           '4', '8', 's', 'p', 'P', 
           '*', 'h', 'H', '+', 'x', 
           'X', 'D', 'd', '|', '_'
          ]
fig, axes = plt.subplots(5, 5)
axes = axes.reshape(25).squeeze()

for i, marker in enumerate(markers):
    axes[i].plot([1], marker, markersize=20) # 标记
    axes[i].set_title(marker)
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    
fig.tight_layout()
plt.show()
```
