本文介绍一些长方体的画法。

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Heiti TC']
````

# `barh`
水平柱状图。
```python
barh(y, width, height=0.8, left=None, *, align='center', data=None, **kwargs)
```
`bar` 中的 `x`、`height` 和 `width` 分别对应 `barh` 里的 `y`、`width` 和 `height`，其它参数与 `bar` 相同，不再赘述。

````python
colors = ['red', 'green', 'blue', 'pink', 'yellow']
y_pos = np.arange(len(colors))
vals = 1 + np.random.rand(len(colors))

fig, ax = plt.subplots()

ax.barh(y_pos, width = vals, color = colors)
ax.set_yticks(range(len(colors)), colors)
ax.set_title('水平柱状图')

plt.show()
````

# `broken_barh`
绘制一系列水平位置相同的矩形。
```python
broken_barh(xranges, yrange, *, data=None, **kwargs)
```
## `xranges` & `yranges`
- `xranges`：元组序列（(左侧 x 坐标, 宽度)）
- `yranges`元组（(y 轴下侧坐标，高度)）

````python
fig, ax = plt.subplots()

ax.broken_barh(xranges = ((1, 1), (3, 2)),
               yrange = (1, 1))
ax.set_title('破碎水平柱状图')

plt.show()
````

## `**kwargs`
接受 `matplotlib.collections.PolyCollection` 的参数。

# `axhspan`
绘制一个水平矩形。
```python
axvspan(xmin, xmax, ymin=0, ymax=1, **kwargs)
```
## `ymin` & `ymax`
浮点数。矩形的上沿和下沿的坐标。
## `xmin` & `xmax`
[0, 1] 之间的浮点数，矩形的左边界和右边界距画布左右边界的百分比，与坐标无关。

````python
fig, ax = plt.subplots()

ax.axhspan(ymin = 0, ymax = 1, xmin = 0.1, xmax = 0.9)
ax.set_title('水平矩形')
ax.set_ylim((-1, 2))

plt.show()
````

# `**kwargs`
接受 `matplotlib.patches.Polygon` 参数。

# `axvspan`
绘制一个竖直矩形。
```python
axvspan(xmin, xmax, ymin=0, ymax=1, **kwargs)
```
## `xmin` & `xmax`
浮点数。矩形的左侧和右侧的坐标。
## `ymin` & `ymax`
[0, 1] 之间的浮点数，矩形的上沿和下沿距画布上下边界的百分比，与坐标无关。

````python
fig, ax = plt.subplots()

ax.axvspan(xmin = 0, xmax = 1, ymin = 0.1, ymax = 0.9)
ax.set_title('竖直矩形')
ax.set_xlim((-1, 2))

plt.show()
````

# `hist2d`
绘制一个 2D 直方图，与热力图的原理大致相同。绘制函数与直方图大致相同。与 `hist` 和 `hexbin` 相同的参数不再赘述。
```python
hist2d(x, y, bins=10, range=None, density=False, weights=None, cmin=None, cmax=None, *, data=None, **kwargs)
```
## `cmin` & `cmax`
计数以后在 `[cmin, cmax]` 之外的箱体将不会显示。

````python
np.random.seed(1)
x = [np.random.randint(0, 100) for _ in range(1000)]
y = [np.random.randint(0, 100) for _ in range(1000)]

fig, (ax1, ax2) = plt.subplots(2)
ax1.hist2d(x, y, bins = 20)
ax1.set_title('')

ax2.hist2d(x, y, bins = 20, cmin = 5, cmax = 10)
ax2.set_title('')

plt.show()
````

# `stairs`
阶梯式矩形（有填充）或线段（没有填充）。
```python
stairs(values, edges=None, *, orientation='vertical', baseline=0, fill=False, data=None, **kwargs)
```
## `values`
数组，台阶的高度。

````python
values_ = [1, 3, 4, 6]
edges_ = [2, 5, 6, 9, 10]

fig, ax = plt.subplots()

ax.stairs(values = values_)
ax.set_title('台阶图')
ax.set_ylim((-1, 8))
fig.savefig('pics/1-12-6.png')
plt.show()
````

## `edges`
数组，长度为 `len(values) + 1`。每个台阶边缘的坐标。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.stairs(values = values_)
ax1.set_title('edges = [0, 1, 2, 3, 4]')

ax2.stairs(values = values_, edges = edges_)
ax2.set_title('edges = [2, 5, 6, 9, 10]')

fig.tight_layout()
plt.show()
````

## `orientation`
在 `'vertical'` 和 `'horizontal'` 里选择，默认为 `'vertical'`。`'vertical'` 是指 台阶的高度是对应 y 轴而言的。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.stairs(values = values_)
ax1.set_title("orientation = 'vertical'")

ax2.stairs(values = values_, orientation = 'horizontal')
ax2.set_title("orientation = 'horizontal'")

fig.tight_layout()
plt.show()
````

## `fill`
布尔值，默认为 `False`。为 `True` 则填充台阶。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.stairs(values = values_)
ax1.set_title("fill = False")

ax2.stairs(values = values_, fill = True)
ax2.set_title("fill = True")

fig.tight_layout()
plt.show()
````

## `baseline`
浮点数或浮点数数组，默认为 0。台阶底部的高度。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.stairs(values = values_)
ax1.set_title("orientation = 'vertical'")

ax2.stairs(values = values_, baseline = [0, 1, 2, 3])
ax2.set_title("orientation = 'horizontal'")
ax2.set_ylim((-2, 8))

fig.tight_layout()
fig.savefig('pics/1-12-10.png')
plt.show()
````
