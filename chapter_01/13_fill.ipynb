本文介绍两个填充的函数。

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Heiti TC']
````

# `fill_between`
这个函数填充两条折线之间的水平空间。
```python
fill_between(x, y1, y2=0, where=None, interpolate=False, step=None, *, data=None, **kwargs)
```

## `x` & `y1` & `y2`
- `x`：数组。折线的 x 轴坐标；
- `y1`：与`x` 长度相同的数组或标量。第一条折线的 y 轴坐标。
- `y2`：与`x` 长度相同的数组或标量，默认为 0。第一条折线的 y 轴坐标。

````python
x = [1, 2, 3, 5]
y1 = [1, 2, 3, 4]
y2 = [2, 4, 6, 5]

fig, ax = plt.subplots()

ax.fill_between(x, y1, y2)
ax.set_title('原始图像')

plt.show()
````

## `where`
布尔值数组，默认为 `None`。对于参数数组索引 `i`，如果 `where[i] == True and where[i+1] == True`，则 `x[i]` 和 `x[i+1]` 之间的区域被填充。

````python
x = np.array([1, 2, 4, 5])
y1 = np.array([0.8, 0.8, 0.2, 0.2])
y2 = np.array([0, 0, 1, 1])

fig, (ax1, ax2) = plt.subplots(2)

ax1.fill_between(x, y1, y2)
ax1.set_title('where = None')

ax2.fill_between(x, y1, y2, where = (y1 > y2))
ax2.fill_between(x, y1, y2, where = (y1 < y2))
ax2.set_title('where = (y1 > y2)')

fig.tight_layout()
plt.show()
````

## `interpolate`
布尔值，默认为 `False`。只有当提供了 `where` 参数且两条线相交时才有效。当为 `False` 时，只有当 `x` 的连续区间满足 `where` 条件时才会填充，若区间有相邻的 True 和 False，则不会填充。为 True 时，程序会猜测使用满足 `where` 条件的地方并填充。

````python
x = np.array([1, 2, 4, 5])
y1 = np.array([0.8, 0.8, 0.2, 0.2])
y2 = np.array([0, 0, 1, 1])

fig, (ax1, ax2) = plt.subplots(2)

ax1.fill_between(x, y1, y2, where = (y1 > y2), color = 'lightblue')
ax1.fill_between(x, y1, y2, where = (y1 < y2), color = 'lightgreen')
ax1.set_title('interpolate = False')

ax2.fill_between(x, y1, y2, where = (y1 > y2), 
                 color = 'lightblue', interpolate = True)
ax2.fill_between(x, y1, y2, where = (y1 < y2), 
                 color = 'lightgreen', interpolate = True)
ax2.set_title('interpolate = True')

fig.tight_layout()
plt.show()
````

## `step`
可为 `'pre'`, `'post'`, `'mid'` 或 `'None'`。如果 `x` 中有连续恒定的数值，使用 `'pre'` 则 y 值使用连续恒定区间左侧的数值；使用 `'post'` 则 y 值使用连续恒定区间右侧的数值；使用 `'mid'` 则连续区间的左侧使用左侧数值，右侧使用右侧的数值；使用 `'None'` 则使用两端的数值。

````python
x = np.array([1, 1, 2, 2])
y1 = np.array([0.8, 0.8, 0.2, 0.2])
y2 = np.array([0, 0, 1, 1])

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4)

ax1.fill_between(x, y1, y2, step = 'pre')
ax1.set_title("step = 'pre'")

ax2.fill_between(x, y1, y2, step = 'post')
ax2.set_title("step = 'post'")

ax3.fill_between(x, y1, y2, step = 'mid')
ax3.set_title("step = 'mid'")

ax4.fill_between(x, y1, y2)
ax4.set_title("step = 'None'")

fig.tight_layout()
fig.savefig('pics/1-13-4.png')
plt.show()
````

## `**kwargs`
接受 `matplotlib.collections.PolyCollection` 的参数。
