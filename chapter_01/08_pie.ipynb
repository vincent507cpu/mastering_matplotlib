饼图的每个部分对应数组中的一个值，扇形面积为 `x / sum(x)`。扇形从 x 轴开始按逆时针显示。

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Heiti TC']

data = np.array([1, 2., 3, 4])
````

````python
fig, ax = plt.subplots()

ax.pie(x = data)
plt.show()
````

```python
pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, normalize=True, hatch=None, data=None)
```

# 图像属性
## `x`，`data`
数据源，一维数组，不再赘述。

## `explode`
`None` 或者与数据长度相等的一维数组，对应部分向外位移。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('explode = None')

exp = [0, 0, .1, .2]
ax2.pie(data, explode = exp)
ax2.set_title(f'explode = {exp}')

plt.show()
````

# `hatch`
字符串或与数据长度相同的列表，默认为 `None`。图像的填充纹理。

````python
fig, (ax1, ax2, ax3) = plt.subplots(nrows = 3, figsize = (4, 12))

ax1.pie(data)
ax1.set_title('colors = None')

hatch = '*'
ax2.pie(data, hatch = hatch)
ax2.set_title(f'hatch = {repr(hatch)}')

hatches = ['\\', '|', '-', 'o']
ax3.pie(data, hatch = hatches)
ax3.set_title(f'hatch = {hatches}')

plt.show()
````

# `colors`
None 或列表，默认为 `None`。各个部分的填充颜色。如果为 None，则使用当前颜色设置。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('colors = None')

colors = ['r', 'b', 'g', 'y']
ax2.pie(data, colors = colors)
ax2.set_title(f'colors = {colors}')

plt.show()
````

# `shadow`
布尔值，默认为 False。True 为饼图添加阴影。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('shadow = False')

ax2.pie(data, shadow = True)
ax2.set_title(f'shadow = True')

plt.show()
````

# `frame`
布尔值，默认为 False。True 则为添加坐标轴。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('frame = False')

ax2.pie(data, frame = True)
ax2.set_title(f'frame = True')

plt.show()
````

# `center`
元组，默认为 (0, 0)。饼图的圆心坐标。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data, frame = True)
ax1.set_title('center = (0, 0)')
ax1.set_axis_on()

ax2.pie(data, frame = True, center = (1, 1))
ax2.set_title(f'center = (1, 1)')

plt.show()
````

# `counterclock`
布尔值，默认为 True。True 则以逆时针顺序排列饼图各部分。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('counterclock = True')

ax2.pie(data, counterclock = False)
ax2.set_title('counterclock = False')

plt.show()
````

# `startangle`
浮点数，默认为 0.0。若不为 0，以逆时针顺序方向旋转图像。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('startangle = 0')

ax2.pie(data, startangle = 90)
ax2.set_title('startangle = 90')

plt.show()
````

# `radius`
浮点数，默认为 1。饼图的半径。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('radius = 1')

ax2.pie(data, radius = 1.5)
ax2.set_title(f'radius = 1.5')

plt.show()
````

# `wedgeprops`
字典，接受 `matplotlib.patches.Wedge` 的参数，默认为 None。扇形实体的属性。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('wedgeprops = None')

wedgeprop = {'linewidth':3, 'edgecolor':'k'}
ax2.pie(data, wedgeprops = wedgeprop)
ax2.set_title("wedgeprops = {'linewidth':3, 'edgecolor':'k'}")

plt.show()
````

# `normalize`
布尔值，默认为 True。True 则对所有数据进行归一化，使 sum(x) == 1。若为 False，当 sum(x) <= 1 时，画一张不完整的饼图；sum(x) >= 1 时抛出异常。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('autopct = None')

ax2.pie(data[:3] / sum(data), autopct = '%.1f%%', normalize = False)
ax2.set_title(f'colors = {colors}')

plt.show()
````

# 文字属性
## `labels`
字符串列表或 None。在对应扇形实体的外侧做标记。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('labels = None')

lbs = ['a', 'b', 'c', 'd']
ax2.pie(data, labels = lbs)
ax2.set_title(f'labels = {lbs}')

plt.show()
````

# `labeldistance`
浮点数，默认为 1.1。标签与扇形实体的距离。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

lbs = ['a', 'b', 'c', 'd']
ax1.pie(data, labels = lbs)
ax1.set_title('labeldistance = 1.1')

ax2.pie(data, labels = lbs, labeldistance = 1.5)
ax2.set_title('labeldistance = 1.5')

plt.show()
````

# `rotatlabels`
布尔值，默认为 False。为 True 则旋转标签，使之与对应扇形实体的方向一致。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('rotatelabels = False')

lbs = ['a', 'b', 'c', 'd']
ax2.pie(data, labels = lbs, rotatelabels = True)
ax2.set_title('rotatelabels = True')

plt.show()
````

# `autopct`
None 或字符串或函数。若为字符串，则在扇形实体内打印格式化后的百分比；若为函数，则打印调用函数后的百分比。

````python
def format_str(s):
    return f'{s / 100:.1%}'

fig, (ax1, ax2, ax3) = plt.subplots(nrows = 3, figsize = (4, 12))

ax1.pie(data)
ax1.set_title('autopct = None')

ax2.pie(data, autopct = '%.1f%%')
ax2.set_title(f"autopct = '%.1f%%'")

ax3.pie(data, autopct = format_str)
ax3.set_title(f"autopct = format_str")
fig.savefig('pics/1-8-18')
plt.show()
````

# `pctdistance`
浮点数，默认值为 0.6。百分比距圆心的百分比。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data, autopct = '%.1f%%')
ax1.set_title('pctdistance = 0.6')

ax2.pie(data, autopct = '%.1f%%', pctdistance = 1.0)
ax2.set_title('pctdistance = 1.0')

plt.show()
````

# `textprops`
字典，默认为 None。标签和百分比字符的属性。

````python
fig, (ax1, ax2) = plt.subplots(nrows = 2, figsize = (4, 8))

ax1.pie(data)
ax1.set_title('textprops = None')

textprop = {'fontsize':20, 'color':'darkgreen'}
lbs = ['a', 'b', 'c', 'd']
ax2.pie(data, autopct = '%.1f%%', labels=lbs,
        textprops=textprop)
ax2.set_title("textprops = {'fontsize':20, 'color':'darkgreen'}")

plt.show()
````
