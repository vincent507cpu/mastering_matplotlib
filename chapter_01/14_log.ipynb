对数曲线与 `plot` 的语法非常类似，只不过 x 或 y 或全部坐标轴变成了对数坐标轴。

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Heiti TC']

x = np.arange(100)
y = np.arange(100) + 1
````

# `loglog`
与 `plot` 的语法几乎完全相同，但是 x 轴、y 轴均为对数坐标轴。
```python
loglog([x], y, [fmt], data=None, **kwargs)
loglog([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
```

````python
fig, ax = plt.subplots()

ax.loglog(x, y)
ax.set_title('loglog')
fig.savefig('pics/1-14-1.png')
plt.show()
````

## `base`
浮点数，默认为 10。对数的底。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.loglog(x, y)
ax1.set_title('base = 10')

ax2.loglog(x, y, base = 2)
ax2.set_title('base = 2')

fig.tight_layout()
plt.show()
````

## `subs`
数组，默认为 `None`。次要刻度的位置。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.loglog(x, y)
ax1.set_title("subs = 'None'")

ax2.loglog(x, y, subs = [])
ax2.set_title("subs = []")

fig.tight_layout()
plt.show()
````

## `nonpositive`
在 'mask' 和 'clip'中选择， 默认为 'clip'。对负数的处理方法。'mask' 则负数不参与绘图，'clip' 则将负数变成一个小正数。

````python
x_ = np.arange(100) - 50
y_ = np.arange(100)

fig, (ax1, ax2) = plt.subplots(2)

ax1.loglog(x_, y_)
ax1.set_title('base = 10')

ax2.loglog(x_, y_, nonpositive = 'mask')
ax2.set_title('base = 2')

fig.tight_layout()
plt.show()
````

# `semilogx`
只将 x 轴变成对数坐标轴，参数与 `loglog` 完全一致。
```python
semilogx([x], y, [fmt], data=None, **kwargs)
semilogx([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
```

````python
fig, ax = plt.subplots()

ax.semilogx(x, y)
ax.set_title('semilogx')

plt.show()
````

```python
semilogy([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
```

````python
fig, ax = plt.subplots()

ax.semilogy(x, y)
ax.set_title('semilogy')

plt.show()
````
