# `text`
```python
text(x, y, s, fontdict=None, **kwargs)
```
- `x` & `y`，`s`
`s` 是字符串，其左下角的坐标是 (x, y)。

````python
fig, ax = plt.subplots()

ax.text(0.5, 0.5, 'hello world')
ax.axvline(0.5, lw = 2, alpha = 0.5)
ax.axhline(0.5, lw = 2, alpha = 0.5)

plt.show()
````

- `fontdict`
字典，默认为 `None`。可覆盖默认设置（保存在 `plt.reParams` 中）。

- `**kwargs`
也可以接受 `matplotlib.text.Text` 的参数。参数很多，请查阅官方文档。举几个例子：

````python
fig, ax = plt.subplots()

ax.text(0.5, 0.5, 'hello world', 
        alpha = 0.5, # 透明度
        color = 'red', # 颜色
        fontsize = 20, # 尺寸
        rotation = 45) # 逆时针旋转的角度
ax.set_title("alpha = 0.5, color = 'red', fontsize = 20, rotation = 45")
fig.savefig('../pics/3-2-2.png')
plt.show()
````
