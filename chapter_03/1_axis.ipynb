````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Heiti TC']

y = np.sin(np.arange(100))
````

本文将讨论如何自定义坐标轴。由于 x 轴和 y 轴的函数基本相同，本文以 y 轴为示范。

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (2, 3))

ax.minorticks_on()
# ax.set_ylim(0, 100)
ax.set_ylabel('y axis label')
ax.set_yticks([0., 0.5, 1.], [0, 50, 100])
ax.set_yticks(np.arange(0, 1, 0.1),
              np.arange(0, 100, 10), 
              minor = True)
ax.tick_params(which = 'major', width = 2.0, length = 20, color = 'red')
ax.tick_params(which = 'minor', width = 2.0, length = 5, color = 'blue')

ax.spines[["top", "bottom", "right"]].set_visible(False)
ax.set_xticks([])
# ax.spines.left.set_position
fig.tight_layout()
plt.savefig('../pics/3-1-1.png')
plt.show()
````

上图所示是一根 y 轴。
- 0 到 100 的长竖线是 y 坐标轴，叫做“脊”（spine）；
- 0、50、100 右侧的长红色横线是“主刻度”（major tick）；
- 10 到 40、60 到 90 右侧的短蓝色横线是“副刻度”（minor tick）；
- 数字是刻度标记（tick label）；
- 左侧的字符串“y axis label”是“y 轴标记”（y axis label）。

# `set_xlim` & `set_ylim`
设置 x 轴或 y 轴的上下界。
```
Axes.set_ylim(bottom=None, top=None, *, emit=True, auto=False, ymin=None, ymax=None)
```
- `bottom` & `top` & `ymin` & `ymax`
浮点数数组。设置 x 轴或 y 轴的上下界。`bottom` 和 `ymin`、`ymax` 和 `top` 不能同时共存。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_title('bottom = None, top = None')

ax2.plot(y)
ax2.set_ylim(bottom = -2, top = 2)
ax2.set_title('bottom = -2, top = 2')

fig.tight_layout()
plt.show()
````

- `emit`
布尔值，默认为 `True`。是否将限制更改通知观察者。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_ylim(bottom = -2, top = 2)
ax1.set_title('emit = True')

ax2.plot(y)
ax2.set_ylim(bottom = -2, top = 2, emit = False)
ax2.set_title('emit = False')

fig.tight_layout()
plt.show()
````

- `auto`
布尔值，默认为 `False`。是否开启 x 轴或 y 轴的自动缩放。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_ylim(bottom = -2, top = 2)
ax1.set_title('auto = False')

ax2.plot(y)
ax2.set_ylim(bottom = -2, top = 2, auto = True)
ax2.set_title('auto = True')

fig.tight_layout()
plt.show()
````

# `set_xlabel` & `set_ylabel`
设置 x 轴或 y 轴的标签。
```python
set_ylabel(ylabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)
```

- `ylabel`
字符串。x 轴或 y 轴的标签。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_ylabel(None)
ax1.set_title('ylabel = None')

ax2.plot(y)
ax2.set_ylabel(ylabel = "y value")
ax2.set_title("y value")

fig.tight_layout()
plt.show()
````

- `fontdict` & `**kwargs`
字典。控制文本的字体、大小、颜色、角度等属性，接受 `matplotlib.text.Text` 的参数。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_ylabel(ylabel = "y value")
ax1.set_title('fontdict = None')

ax2.plot(y)
ax2.set_ylabel(ylabel = "y value", fontdict = {'fontsize':12, 'color':'red'}, rotation=45)
ax2.set_title("fontdict = {'fontsize':12, 'color':'red'}, rotation=45")

fig.tight_layout()
plt.show()
````

- `labelpad`
浮点数，默认值为 `rcParams["axes.labelpad"]`（4.0）。标签与画框边缘之间的间隙。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_ylabel(ylabel = "y value")
ax1.set_title('labelpad = 4.0')

ax2.plot(y)
ax2.set_ylabel(ylabel = "y value", labelpad = 50.0)
ax2.set_title("labelpad = 50.0")

fig.tight_layout()
plt.show()
````

- `loc`
'bottom'、'center' 或 'top'，默认值为 `rcParams["yaxis.labellocation"]`（'center'）。标签的位置。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_ylabel(ylabel = "y value")
ax1.set_title("loc = 'center'")

ax2.plot(y)
ax2.set_ylabel(ylabel = "y value", loc = 'top')
ax2.set_title("loc = 'top'")

fig.tight_layout()
plt.show()
````

# `set_xscale` & `set_yscale`
```python
set_yscale(value, **kwargs)
```
- `value`
"linear"、"log"、"symlog"、"logit" 或自定义缩放方法，默认为 `linear`。轴上值的缩放方法。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_title('ylim = None')

ax2.plot(y)
ax2.set_yscale('log')
ax2.set_title('yscale = "log"')

fig.tight_layout()
plt.show()
````

- `**kwargs`
与上面的 `value` 对应，比如 `value = 'linear'` 时，此处接受 `matplotlib.scale.LinearScale` 的参数。

# `set_xticks` & `set_yticks`
```python
set_yticks(ticks, labels=None, *, minor=False, **kwargs)
```
- `ticks` & `labels`
字符串数组和浮点数数组。这两个参数一般在一起使用，一个定义了刻度的位置，一个定义了对应位置的值。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_title('ticks = None, labels = None')

ax2.plot(y)
ax2.set_yticks(ticks = np.arange(-1, 1, 0.4), labels = np.arange(-1, 1, 0.4))
ax2.set_title('ticks = np.arange(-1, 1, 0.4), labels = np.arange(-1, 1, 0.4)')

fig.tight_layout()
plt.show()
````

- `minor`
布尔值，默认为 `False`。若为 `True`，则词函数里的参数会影响次要刻度；若为 `False`，则词函数里的参数会影响主要刻度。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_yticks(ticks = np.arange(-1, 1, 0.4),
               labels = np.arange(-1, 1, 0.4))
ax1.set_title('minor = False')

ax2.plot(y)
ax2.set_yticks(ticks = np.arange(-1, 1, 0.25), 
               labels = np.arange(-1, 1, 0.25),
               color='red', minor = True)
ax2.set_title('minor = True')

fig.tight_layout()
plt.show()
````

# `minorticks_off` & `minorticks_on`
没有参数。是否显示次要刻度。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_title('minorticks_off')

ax2.plot(y)
ax2.minorticks_on()
ax2.set_title('minorticks_on')

fig.tight_layout()
plt.show()
````

# `tick_params`
```python
tick_params(axis='both', **kwargs)
```
- `axis`
'x'、'y' 或 'both'，默认为 'both'。将属性应用在哪个轴上。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.tick_params(color = 'red')
ax1.set_title("axis = 'both', color = 'red'")

ax2.plot(y)
ax2.tick_params(axis = 'y', color = 'red')
ax2.set_title("axis = 'y', color = 'red'")

fig.tight_layout()
plt.show()
````

- `which`
'major', 'minor', 'both'，默认为 'major'。将属性应用在主要刻度还是次要刻度上。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.minorticks_on()
ax1.tick_params(color = 'red', width=2)
ax1.set_title("which = 'major', color = 'red', width=2")

ax2.plot(y)
ax2.minorticks_on()
ax2.tick_params(which = 'both', color = 'red', width=2)
ax2.set_title("which = 'both', color = 'red', width=2")

fig.tight_layout()
plt.show()
````

- `**kwargs`
设置刻度的长度、颜色等属性。

# `set_xmargin` & `set_ymargin`
```python
set_ymargin(m)
```
设置数据到 y 轴之间的填充。
- `m`
大于 -0.5 的浮点数。$m$ 乘以数据跨度的值会被加在数据的两端。若 $m$ 为负，则图像将被裁剪。

````python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(y)
ax1.set_title('ymargin = 0.0')

ax2.plot(y)
ax2.set_ymargin(1.0)
ax2.set_title('ymargin = 1.0')

fig.tight_layout()
plt.show()
````

# `ax.spines`
一幅图默认有 4 个脊（`up`，`bottom`，`left`，`right`）。如果想控制其中的任意一个，既可以使用类方法的形式（`ax.spines.left`），也可以使用 pandas 选取特征的形式（`ax.spines['left']` 或 `ax.spines[['left']]`）。如果希望同时选择多个，则可使用后一个方法（`ax.spines[['left', 'top']]`）或循环体。若同时控制所有脊，可直接调用 `ax.spines`。脊比较重要的属性有：
- `set_alpha`
- `set_color`
- `set_edgecolor` 或 `set_ec` 或 `set_color`
- `set_fc` 或 `set_facecolor`
- `set_joinstyle`
- `set_ls` 或 `set_linestyle`
- `set_lw` 或 `set_linewidth`
- `set_position`
- `set_visible`

````python
spines = ['top', 'left', 'bottom', 'right']
colors = ['red', 'blue', 'green', 'black']

fig, ax = plt.subplots()
for side, color in zip(spines, colors):
    ax.spines[side].set_color(color)
    ax.spines[side].set_lw(2)
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
````
